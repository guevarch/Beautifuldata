```python
import numpy as np

# Set the number of rows and columns
num_rows = 100
num_columns = 2

# Create a time variable
t = np.linspace(0, 10*np.pi, num_rows)

# Create the spiral dataset
x = np.cos(t) * t
y = np.sin(t) * t
z = np.linspace(0, 10, num_rows)

# Combine the x, y, and z coordinates into a single ndarray
data = np.column_stack((x, y, z))
# Set the number of rows and column

import plotly.graph_objects as go

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=data[:, 0],  # x-coordinates
    y=data[:, 1],  # y-coordinates
    z=data[:, 2],  # z-coordinates
    mode='markers',
    marker=dict(
        size=3,
        color=data[:, 2],  # set color to z values
        colorscale='Viridis',  # choose a colorscale
        opacity=0.8
    )
)])

# Set the title and axis labels
fig.update_layout(scene=dict(
    xaxis_title='X',
    yaxis_title='Y',
    zaxis_title='Z'),
    title="3D Spiral Plot"
)

# Show the plot
fig.show()
```
