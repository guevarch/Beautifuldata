```python
import pandas as pd

df = pd.read_csv('Resources/token_report2.csv')
df
```

```python
df1 = pd.read_csv('Resources/report.csv')
df1
```

```python
df1 = df1.rename(columns={'Account Name': 'Redeemer Name'})
merged_df = df1.merge(df, on='Redeemer Name')
merged_df.to_csv('merged_data.csv', index=False)

merged_df
```

```python
similar_names_count = df1['Redeemer Name'].isin(df['Redeemer Name']).sum()
similar_names_count
```

```python
# only count unique
unique_enrollment_count = merged_df['Program Enrollment: Enrollment ID'].nunique()
unique_enrollment_count
filtered_df = merged_df.drop_duplicates(subset='Program Enrollment: Enrollment ID')
filtered_df.to_csv('filtered_df.csv', index=False)

```
